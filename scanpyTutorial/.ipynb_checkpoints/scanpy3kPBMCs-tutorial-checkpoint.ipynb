At the end of this tutorial, we will output the derived p-values of differential expression between the cell clusters. Then, we change over to R to read those p-values and use scPPIN to identify the functional modules for each cluster.

```python
import numpy as np
import pandas as pd
import scanpy as sc
```

```python
sc.settings.verbosity = 3             # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_versions()
sc.settings.set_figure_params(dpi=80)
```

```python
results_file = './write/pbmc3k.h5ad'  # the file that will store the analysis results
```

```python
adata = sc.read_10x_mtx(
    './data/filtered_gene_bc_matrices/hg19/',  # the directory with the `.mtx` file
    var_names='gene_symbols',                  # use gene symbols for the variable names (variables-axis index)
    cache=True)     
```

```python
adata.var_names_make_unique()  # this is unnecessary if using `var_names='gene_ids'` in `sc.read_10x_mtx`
```

## Preprocessing 

```python
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.filter_genes(adata, min_cells=3)
```

```python
mito_genes = adata.var_names.str.startswith('MT-')
# for each cell compute fraction of counts in mito genes vs. all genes
# the `.A1` is only necessary as X is sparse (to transform to a dense array after summing)
adata.obs['percent_mito'] = np.sum(
    adata[:, mito_genes].X, axis=1).A1 / np.sum(adata.X, axis=1).A1
# add the total counts per cell as observations-annotation to adata
adata.obs['n_counts'] = adata.X.sum(axis=1).A1
```

```python
adata = adata[adata.obs.n_genes < 2500, :]
adata = adata[adata.obs.percent_mito < 0.05, :]
```

```python
sc.pp.normalize_total(adata, target_sum=1e4)
```

```python
sc.pp.log1p(adata)
```

```python
adata.raw = adata
```

```python
sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5)
```

```python
adata = adata[:, adata.var.highly_variable]
```

```python
sc.pp.regress_out(adata, ['n_counts', 'percent_mito'])
```

```python
sc.pp.scale(adata, max_value=10)
```

```python
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=40)
```

```python
sc.tl.umap(adata)
```

```python
sc.tl.louvain(adata,resolution=1.2)
```

```python
sc.pl.umap(adata, color=['louvain', 'CST3', 'NKG7'],legend_loc='on data')
```

```python
# the scanpy tutorial provides us with the cell type names
# Be aware that the order is slightly reordered from the tutorial
new_cluster_names = [
    'CD4 T', 
    'CD14 Monocytes',
    'CD8 T',
    'B', 
    'FCGR3A Monocytes',
    'NK', 
    'Dendritic', 
    'Megakaryocytes']
adata.rename_categories('louvain', new_cluster_names)
```

```python
sc.pl.umap(adata, color='louvain',legend_loc='on data', title='', frameon=False, save='.pdf')
```

```python
adata.write(results_file)
```

## Find Marker Genes

```python
# note that we set the n_genes for the DEG (rank_genes_groups) to the maximal possible number, such that we test all genes
numGenes = 2638
sc.tl.rank_genes_groups(adata, 'louvain', method='wilcoxon',n_genes=numGenes,corr_method='bonferroni')
sc.pl.rank_genes_groups(adata, n_genes=25, sharey=False, save='.pdf')
```

```python
# we now save all the p-values as a (Celltype)x(Genes) matrix
# we use a csv

DEG_matrix = np.shape(np.array(adata.uns['rank_genes_groups']['pvals'].tolist()))

np.savetxt('degPBMC.txt',DEG_matrix,delimiter=',')

#adata.uns['rank_genes_groups']['pvals'].tofile('degPBMC.txt',sep=",")
```
